```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import time
import glob
```

```python
path = "G:/darknet-master/darknet-master/build/darknet/x64/data/train_bdd"
```

```python
label_set = ['car', 'bus', 'person', 'bike', 'truck', 'motor', 'train', 'rider', 'traffic sign', 'traffic light']
```

```python
d = dict(zip(label_set,[0 for _ in range(len(label_set))]))
```

```python
for file in glob.glob(path+'/*.txt'):
    with open(file,'r',encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            l_res = []
            l = line.split()
            ind = int(l[0])
            d[label_set[ind]] += 1
            
    

    
print("work done!")
```

```python
source_data = d

for a, b in source_data.items():
    plt.text(a, b + 0.05, '%.0f' % b, ha='center', va='bottom', fontsize=11)  


x_axis = tuple(source_data.keys())
y_axis = tuple(source_data.values())
plt.bar(x_axis, y_axis)  

plt.xlabel("classes") 
plt.xticks(rotation=45)
plt.ylabel("number of objects") 
plt.title("bdd_statistics")  
plt.ylim(0,800000) 
plt.savefig("bdd_stat.png")
plt.show()
```
